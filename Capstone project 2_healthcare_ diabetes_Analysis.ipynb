```python
features = data.iloc[:,[1,2,3,4,5,6,7]].values
label = data.iloc[:,8].values
```

```python
#train_test_split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,label, test_size = 0.2,random_state = 0) 
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(label, model.predict(features))
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(label, model.predict(features)))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

probs = model.predict_proba(X_test)

# keep probabilities for the positive outcome only
probs = probs[:,1]

# caluculate AUC
auc = roc_auc_score(y_test,probs)
print("AUC: %.3F" % auc)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(X_train,y_train)
```

```python
model3.score(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```

```python
# Random forest

from sklearn.ensemble import RandomForestClassifier
model4 = RandomForestClassifier()
model4.fit(X_train,y_train)
```

```python
model4.score(X_train,y_train)
```

```python
model4.score(X_test,y_test)
```

```python
# Support Vector Classifier

from sklearn.svm import SVC
model5= SVC(kernel = "rbf", gamma = 'auto')
model5.fit(X_train,y_train)
```

```python
model5.score(X_train,y_train)
```

```python
model5.score(X_test,y_test)
```

```python
# Knn

from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=7, metric='minkowski', p=2)
model2.fit(X_train,y_train)
```

```python
model2.score(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

probs = model2.predict_proba(X_test)

# keep probabilities for the positive outcome only
probs = probs[:,1]

# caluculate AUC
auc = roc_auc_score(y_test,probs)
print("AUC: %.3F" % auc)

#calculate the roc curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
plt.plot([0,1],[0,1], linestyle="--")
#plot the roc curve for the model
plt.plot(fpr,tpr,marker='.')
```

```python
#  precision recall curve for decision tree classifier

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
from matplotlib import pyplot


probs = model2.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:,1]

# caluculate scores
precision,recall,thresholds = precision_recall_curve(y_test,probs)

plt.plot(recall,precision,marker='.')
```
